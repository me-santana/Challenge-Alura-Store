### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
# Calculando o faturamento total de cada loja

faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Criando uma lista para armazenar os resultados

faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

# Exibindo os resultados de forma organizada

print(f"Faturamento Loja 1: R$ {faturamento_loja1:,.2f}")
print(f"Faturamento Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento Loja 4: R$ {faturamento_loja4:,.2f}")
```

```python
# Gráfico para melhor visualização do faturamento total de cada loja

import matplotlib.pyplot as plt


nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']


plt.figure(figsize=(8, 4)) #tamanho da imagem
plt.bar(nomes_lojas, faturamentos, color=['#6A0DAD', '#9932CC', '#BA55D3', '#D8BFD8'])


plt.title('Faturamento Total por Loja', fontsize=15)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)


plt.show()

```

# 2. Vendas por Categoria

```python
cat_loja1 = loja['Categoria do Produto'].value_counts()
cat_loja2 = loja2['Categoria do Produto'].value_counts()
cat_loja3 = loja3['Categoria do Produto'].value_counts()
cat_loja4 = loja4['Categoria do Produto'].value_counts()

from IPython.display import display

display(cat_loja1)
display(cat_loja2)
display(cat_loja3)
display(cat_loja4)

```

```python
categorias_totais = [cat_loja1.sum(), cat_loja2.sum(), cat_loja3.sum(), cat_loja4.sum()]

import matplotlib.pyplot as plt


nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 4)) #tamanho da imagem
plt.bar(nomes_lojas, categorias_totais, color=['#6A0DAD', '#9932CC', '#BA55D3', '#D8BFD8'])

plt.title('Total de Itens vendidos por Lojas', fontsize=15)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Quantidade de Itens', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)


plt.show()
```

```python
import matplotlib.pyplot as plt

# Minha paleta de cores preferidas (do mais escuro para o mais claro)
cores_roxas = ['#4B0082', '#6A0DAD', '#8A2BE2', '#9370DB', '#BA55D3']

# Criando uma figura com 4 subgráficos (um para cada loja)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Categorias Mais Populares por Loja', fontsize=20, color='#4B0082')


# Loja 1
cat_loja1.head(5).plot(kind='bar', ax=axes[0, 0], color=cores_roxas)
axes[0, 0].set_title('Loja 1')
axes[0, 0].tick_params(axis='x', rotation=45)

# Loja 2
cat_loja2.head(5).plot(kind='bar', ax=axes[0, 1], color=cores_roxas)
axes[0, 1].set_title('Loja 2')
axes[0, 1].tick_params(axis='x', rotation=45)

# Loja 3
cat_loja3.head(5).plot(kind='bar', ax=axes[1, 0], color=cores_roxas)
axes[1, 0].set_title('Loja 3')
axes[1, 0].tick_params(axis='x', rotation=45)

# Loja 4
cat_loja4.head(5).plot(kind='bar', ax=axes[1, 1], color=cores_roxas)
axes[1, 1].set_title('Loja 4')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 3. Média de Avaliação das Lojas


```python
media_loja1 = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

media_total_avaliacoes = [media_loja1, media_loja2, media_loja3, media_loja4]

print(f"Média Loja 1: {media_loja1:.2f}")
print (f"Média da Loja 2: {media_loja2:.2f}")
print(f"Média da Loja 3: {media_loja3:.2f}")
print(f"Média da Loja 4: {media_loja4:.2f}")
```

```python
import matplotlib.pyplot as plt

nome_das_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']


plt.figure(figsize=(8, 4)) #tamanho da imagem
plt.bar(nome_das_lojas, media_total_avaliacoes, color=['#6A0DAD', '#9932CC', '#BA55D3', '#D8BFD8'])


plt.title('Nível de Satisfação dos Clientes por Unidade', fontsize=15)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Avaliação Média (0 a 5)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ylim(0, 5) # O gráfico mostrar de 0 a 5

plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# Contando as vendas de cada produto por loja
prod_loja1 = loja['Produto'].value_counts()
prod_loja2 = loja2['Produto'].value_counts()
prod_loja3 = loja3['Produto'].value_counts()
prod_loja4 = loja4['Produto'].value_counts()

# Exibindo o Top 1 (o mais vendido) de cada loja
print("Mais vendido Loja 1:", prod_loja1.idxmax(), "-", prod_loja1.max(), "vendas")
print("Mais vendido Loja 2:", prod_loja2.idxmax(), "-", prod_loja2.max(), "vendas")
print("Mais vendido Loja 3:", prod_loja3.idxmax(), "-", prod_loja3.max(), "vendas")
print("Mais vendido Loja 4:", prod_loja4.idxmax(), "-", prod_loja4.max(), "vendas")

# Exibindo o menos vendido de cada loja
print("\nMenos vendido Loja 1:", prod_loja1.idxmin(), "-", prod_loja1.min(), "vendas")
print("Menos vendido Loja 2:", prod_loja2.idxmin(), "-", prod_loja2.min(), "vendas")
print("Menos vendido Loja 3:", prod_loja3.idxmin(), "-", prod_loja3.min(), "vendas")
print("Menos vendido Loja 4:", prod_loja4.idxmin(), "-", prod_loja4.min(), "vendas")
```

```python
import matplotlib.pyplot as plt

nomes_produtos = [prod_loja1.idxmax(), prod_loja2.idxmax(), prod_loja3.idxmax(), prod_loja4.idxmax()]
qtd_vendas = [prod_loja1.max(), prod_loja2.max(), prod_loja3.max(), prod_loja4.max()]
nome_das_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8, 4))

# No barh, passamos primeiro os nomes e depois os valores
plt.barh(nome_das_lojas, qtd_vendas, color=['#6A0DAD', '#9932CC', '#BA55D3', '#D8BFD8'])

# Adicionando o nome do produto e a quantidade na frente de cada barra
for i, v in enumerate(qtd_vendas):
    plt.text(v + 1, i, f" {nomes_produtos[i]} ({v} un.)", va='center', fontweight='bold')

plt.title('Produto mais vendido em cada Loja', fontsize=15)
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Lojas')

# Aumenta o limite lateral para o texto do produto não ser cortado
plt.xlim(0, max(qtd_vendas) + 40)

plt.show()
```

# 📊 Relatório Final

### 1. Introdução
Este relatório tem como objetivo analisar o desempenho das quatro unidades da Alura Store para identificar qual delas apresenta os indicadores mais críticos, subsidiando a decisão do Senhor João sobre qual loja deve ser colocada à venda. A análise considerou faturamento, satisfação do cliente, eficiência logística e mix de produtos.

### 2. Desenvolvimento e Análise dos Dados
Após o processamento dos dados, observamos os seguintes pontos de destaque:

- Faturamento: A Loja 4 apresentou o faturamento total mais baixo entre todas as unidades, indicando um volume menor de transações financeiras.

- Satisfação do Cliente: Surpreendentemente, a Loja 1 obteve a menor média de avaliação da rede. Isso indica uma falha grave na experiência de compra ou na qualidade dos produtos entregues por essa unidade.

- Eficiência Logística (Frete): A Loja 1 possui o frete médio mais caro (R$ 34,69), enquanto a Loja 4 é a mais eficiente nesse quesito (R$ 31,28).

- Produtos e Categorias: Embora a Loja 4 venda menos, seus produtos campeões de vendas mantêm um público fiel e satisfeito. Já a Loja 1, apesar de vender categorias populares, não consegue converter isso em avaliações positivas.

### 3. Conclusão e Recomendação Final
Com base nos dados apresentados, minha recomendação técnica para o Senhor João é a venda da Loja 1.


Pois embora a Loja 4 tenha um faturamento menor, ela opera de forma saudável: tem o frete mais barato da rede e clientes satisfeitos. Por outro lado, a Loja 1 representa um risco para a marca Alura Store. Ela combina os dois piores indicadores operacionais: o custo de entrega mais alto e o maior índice de insatisfação dos clientes. Manter a Loja 1 significa gastar mais para entregar um serviço que está afastando os consumidores. Portanto, a venda da Loja 1 traria um alívio imediato nos custos logísticos e protegeria a reputação da empresa.
